```python

# 加入指定目標路徑
import sys, os
sys.path.append(os.pardir)

import numpy as np

# 小型讀取圖片程式
import pickle

# 注意這個 import 順序要放在 sys 後面
# dataset.mnist 相當於執行 dataset 目錄底下的 mnist.py
from dataset.mnist import load_mnist

# get_data函式化
def get_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(normalize=True, flatten=True, one_hot_label=False)
    return x_test, t_test

# 這裡檔案必須跟程式放在同一資料夾或子資料夾
# rb = 讀取成 2 進位
# 參考資料 https://www.cnblogs.com/ymjyqsx/p/6554817.html
# 這個檔案是最佳值權重，因為還沒有寫相對應程式
def init_network():
    with open("sample_weight.pkl", 'rb') as f:
        network = pickle.load(f)
#         print(network)
    return network

# S 型函數
# 參考 https://baike.baidu.com/item/Sigmoid%E5%87%BD%E6%95%B0/7981407
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# softmax 函數用於算出參數的相對的比例，也就是權重
# 參考 https://www.zhihu.com/question/23765351
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y

# 使用之前的測試資料的類神經網路
# 等同之前的 forward (順向計算)，僅命名不同 
# 必須留意 waiting 值 (矩陣的大小)
# 矩陣檢查看 W1 / W2 / W3
def predict(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)
#     print(W1.shape) 
#     print(W2.shape)
#     print(W3.shape)
    
    return y

x, t = get_data()
network = init_network()

#正確率 , y為輸出值 , p 為取得機率最大值數字
accuracy_cnt = 0
for i in range(len(x)):
    y = predict(network, x[i])
    p= np.argmax(y)
#     print(p,t[i])
    if p == t[i]:
        accuracy_cnt += 1
#         print(accuracy_cnt)
# 正確時累加一次，故顯示  1 2 3 4 ..... 9352


# 除掉訓練的總比數就等於機率
print('Accuracy:'+ str(float(accuracy_cnt)/len(x)))



```
