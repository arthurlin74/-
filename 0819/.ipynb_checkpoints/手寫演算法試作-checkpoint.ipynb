```python
# 參考 0819/04_多維矩陣運算 講義

# 此範例為線性函數
# 除此之外有回歸預測 / 反證 等等用法
# 此範例權重都是隨便給的，資料也只是測試用亂給的
# w = waiting 值
# b = bias 值
# a = 實際權重
# X = 資料
# sigmoid = S型函數
import numpy as np

def sigmoid(x):
    return 1 / (1+np.exp(-x))
print("--------------------------")
print("第一層輸入層")

def identity_function(y):  #沒有特別處理，當作理想狀態
    return (y)

X = np.array([1.0,0.5])
W1 = np.array([[0.1,0.3,0.5],[0.2,0.4,0.6]])
B1 = np.array([0.1,0.2,0.3])

print(X.shape)
print(W1.shape)
print(B1.shape)

A1 = np.dot(X,W1)+B1
print("注意結果 A1 =",A1.shape)
print("其結果顯示 3，代表1*3矩陣，三個輸入元")

Z1 = sigmoid(A1)
print(Z1.shape)

print("--------------------------")
print("第二層，隱藏層")
W2 = np.array([[0.1,0.4],[0.2,0.5],[0.3,0.6]])
print(W2.shape)

B2 = np.array([0.1,0.2])
print(B2.shape)

A2 = np.dot(Z1,W2)+B2
print("注意結果 A2 =",A2)
print("其結果顯示 2 個結果，代表 2 個隱藏層")

Z2 = sigmoid(A2)
print(Z2)

print("--------------------------")

W3 = np.array([[0.1,0.3],[0.2,0.4]])
print(W3.shape)

B3 = np.array([0.1,0.2])
print(B3.shape)

A3 = np.dot(Z2,W3)+B3  #取上一層的輸出與此層的權重再加上補償
print("注意結果 A3 =",A3)
print("其結果顯示 2 個結果，代表 2 個輸出層")

Y = identity_function(A3) #線性關係
print(Y)
```
